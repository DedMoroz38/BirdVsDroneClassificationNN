```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist

(x_train, y_train), (_, _) = mnist.load_data()

image = x_train[403].astype("float32")

def max_pooling_2x2(image):
    height, width = image.shape
    pooled_height = height // 2
    pooled_width = width // 2

    pooled_image = np.zeros((pooled_height, pooled_width))

    for i in range(pooled_height):
        for j in range(pooled_width):
            window = image[2 * i : 2 * i + 2, 2 * j : 2 * j + 2]
            pooled_image[i, j] = np.max(window)

    return pooled_image


pooled_image = max_pooling_2x2(image)
```

```python
plt.figure(figsize=(6, 2))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap="gray")
plt.title(f"Original Image {image.shape}")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(pooled_image, cmap="gray")
plt.title(f"After 2x2 Max-Pooling {pooled_image.shape}")
plt.axis("off")

plt.show()
```
